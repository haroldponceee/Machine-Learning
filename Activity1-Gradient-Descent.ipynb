```python
import statsmodels.formula.api as smf
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
activity1 = pd.read_csv(r'C:\Users\User\Documents\MACHINE LEARNING\Advertising.csv', index_col=0)
activity1.pop('Radio')
activity1.pop('Newspaper')
activity1.head()
```

```python
lm1 = smf.ols(formula='Sales ~ TV', data=activity1).fit() #create fitted model
lm1.params #print the coefficient
```

```python
feature_cols = ['TV']
x = activity1[feature_cols]
y = activity1.Sales

lm2 = LinearRegression()
lm2.fit(x, y)

print(lm2.intercept_)
print(lm2.coef_)
```

```python
sns.pairplot(activity1, x_vars=['TV'], y_vars='Sales', height=7, aspect=0.7, kind='reg')
```

```python
xx = activity1.iloc[:, 0]
yy = activity1.iloc[:, 1]
plt.xlabel('TV')
plt.ylabel('Sales')
plt.scatter(xx, yy, color='red')
plt.show()
```

```python
ypred = lm2.coef_*xx +  lm2.intercept_  #ypred= 0.04753664*XX + 7.032593549127695
plt.scatter(xx, yy, color='red') 
plt.plot([min(xx), max(xx)], [min(ypred), max(ypred)], color='black')
plt.show()
```
